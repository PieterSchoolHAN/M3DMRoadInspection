```python
# dirs
test = r'test\labels'
train = r'train\labels'
valid = r'valid\labels'
labels = ['crack, pothole']
yaml = "data.yaml"
```

```python
import os

def update_labels(directory, to_replace, replace_to):
    """
    Update all label files in the specified directory.
    Changes class 2 to class 1 while leaving other classes unchanged.
    """
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):  # Assuming label files have a .txt extension
            filepath = os.path.join(directory, filename)
            
            with open(filepath, 'r') as file:
                lines = file.readlines()
            
            updated_lines = []
            for line in lines:
                # Split the line into components
                parts = line.split()
                
                # Check each part and update class identifiers
                updated_parts = []
                for part in parts:
                    # Update class if it's a whole number equal to 2
                    if part.isdigit() and part == to_replace:
                        updated_parts.append(replace_to)  # Change class 2 to 1
                    else:
                        updated_parts.append(part)
                
                # Join the updated parts back into a single line
                updated_line = ' '.join(updated_parts)
                updated_lines.append(updated_line)
            
            # Write the updated lines back to the file
            with open(filepath, 'w') as file:
                file.write('\n'.join(updated_lines))
            print(f"Updated {filename}")
```

```python
# Specify the directory containing the label files
dirs = [test, train, valid]
for dir in dirs:
    update_labels(dir, '2', '1')
```

```python
# Adjusting data.yaml
import os

def generate_data_yaml(base_dir, nc, names, output_file="data.yaml"):
    data_structure = {
        "train": os.path.join(base_dir, "train", "images"),
        "val": os.path.join(base_dir, "valid", "images"),
        "test": os.path.join(base_dir, "test", "images"),
        "nc": nc,
        "names": names,
    }
    with open(output_file, 'w') as yaml_file:
        yaml_file.write(f"train: {data_structure['train']}\n")
        yaml_file.write(f"val: {data_structure['val']}\n")
        yaml_file.write(f"test: {data_structure['test']}\n\n")
        yaml_file.write(f"nc: {data_structure['nc']}\n")
        yaml_file.write(f"names: {data_structure['names']}\n")
    
    print(f"{output_file} generated successfully!")

base_dir = ".."
# Generate data.yaml
generate_data_yaml(base_dir, nc=2, names=labels, output_file=yaml)
```
